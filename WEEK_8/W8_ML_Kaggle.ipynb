```python
def find_outliers(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3-q1 
    floor = q1 - 1.5*iqr
    ceiling = q3 + 1.5*iqr

    return ceiling,floor
for col in train_df.columns[:-1]:
    ur,lr=find_outliers(train_df[col])
    print("For: ",col,":")
    d=train_df.shape[0]-train_df[(train_df[col]>=lr) & (train_df[col]<=ur) ].shape[0]
    print("length of outliers data= ",d)
    print("percentage of outliers data= {:0.2f} %".format((d/train_df.shape[0])*100))
    print("__________________________________________________________________________")
```

```python
for col in train_df.columns[:-1]:
    ur,lr=find_outliers(train_df[col])
    train_df[col]=np.where(train_df[col]>ur,ur,train_df[col])
    train_df[col]=np.where(train_df[col]<lr,lr,train_df[col])
plt.figure(figsize=(10,8))
train_df.boxplot(rot=90)
```

```python
df=train_df.copy()
from sklearn.model_selection import train_test_split
X=df.drop("target_class",axis=1)
Y=df["target_class"]
```

```python
Y.value_counts()
```

```python
#! pip install -U imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

smote =SMOTE(sampling_strategy="minority")

x_sm,y_sm= smote.fit_resample(X,Y)
y_sm.value_counts()
```

```python
y_sm.shape
```

```python
x_sm.shape
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(x_sm,y_sm, test_size=0.25 , random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix,ConfusionMatrixDisplay,roc_curve , roc_auc_score,plot_roc_curve
```

```python
svc_rbf = SVC(random_state=0,C=100, gamma=10 , kernel='rbf',probability=True)
svc_rbf.fit(X_train, Y_train)
y_pred_svc_rbf = svc_rbf.predict(X_test)
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(Y_test, y_pred_svc_rbf)))
```

```python
# Check for overfitting and underfitting
print(f"Training Data Score: {svc_rbf.score(X_train, Y_train)}")
print(f"Testing Data Score: {svc_rbf.score(X_test, Y_test)}")
```

```python
print(classification_report(Y_test, y_pred_svc_rbf))
cm2 = confusion_matrix(Y_test, y_pred_svc_rbf)
disp = ConfusionMatrixDisplay(cm2).plot(cmap=plt.cm.Blues)
```

```python
plot_roc_curve(svc_rbf, X_test, Y_test)
```

```python
svc_linear = SVC(random_state=0,C=100,gamma=10,kernel='linear')
svc_linear.fit(X_train, Y_train)
y_pred_svc_linear = svc_linear.predict(X_test)
```

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(Y_test, y_pred_svc_linear)))
```

```python
# Check for overfitting and underfitting
print(f"Training Data Score: {svc_linear.score(X_train, Y_train)}")
print(f"Testing Data Score: {svc_linear.score(X_test, Y_test)}")
```

```python
print(classification_report(Y_test, y_pred_svc_linear))
cm2 = confusion_matrix(Y_test, y_pred_svc_linear)
disp = ConfusionMatrixDisplay(cm2).plot(cmap=plt.cm.Blues)
```
